## 1. SCALING VECTOR

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the original vector
vector = np.array([3, 4])

# Define the scaling matrix
scaled_vector = vector @ np.array([[2, 0],[0, 2]]) # Increase the magnitude by scaling the original vector

# Plot the vector with increased magnitude
plt.figure()
plt.quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Scaled Vector')
plt.text(scaled_vector[0], scaled_vector[1], f'({scaled_vector[0]}, {scaled_vector[1]})', verticalalignment='bottom',
         horizontalalignment='right', fontsize=12, color='r')

# Plot the original vector

plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Original Vector')
plt.text(vector[0], vector[1], f'({vector[0]}, {vector[1]})', verticalalignment='bottom', horizontalalignment='right',
         fontsize=12, color='b')


plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Scaled')
plt.grid()
plt.legend()
plt.show()
```

## 2. ROTATING VECTOR

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the original vector
vector = np.array([3, 4])

# Define the rotating matrix
rotated_vector = vector @ np.array([[-1, 0],[0, 1]]) # Increase the magnitude by scaling the original vector

# Plot the vector with increased magnitude
plt.figure()
plt.quiver(0, 0, rotated_vector[0], rotated_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Rotated Vector')
plt.text(rotated_vector[0], rotated_vector[1], f'({rotated_vector[0]}, {rotated_vector[1]})', verticalalignment='bottom',
         horizontalalignment='right', fontsize=12, color='r')

# Plot the original vector

plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Original Vector')
plt.text(vector[0], vector[1], f'({vector[0]}, {vector[1]})', verticalalignment='bottom', horizontalalignment='right',
         fontsize=12, color='b')


plt.xlim(-5, 6)
plt.ylim(-2, 6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Rotated')
plt.grid()
plt.legend()
plt.show()
```

## 3. RANDOM Transformation

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
vector = np.array([3, 4])

# Define the random matrix
random_matrix = np.array([[2, 3], [4, 1]])

# Apply transformation to the vector
transformed_vector = random_matrix @ vector

# Calculate the dot product of the two vectors
dot_product = np.dot(vector, transformed_vector)

# Calculate the magnitude of each vector
magnitude_vector = np.linalg.norm(vector)
magnitude_scaled_vector = np.linalg.norm(transformed_vector)

# Calculate the cosine of the angle between the vectors using the dot product formula
cos_theta = dot_product / (magnitude_vector * magnitude_scaled_vector)

# Calculate the angle in radians
theta = np.arccos(cos_theta)

# Convert angle to degrees
theta_degrees = np.degrees(theta)

# Plot the vectors
plt.figure()
plt.quiver(0, 0, transformed_vector[0], transformed_vector[1], angles='xy', scale_units='xy', scale=1, color='r',
           label='Rotated and Scaled Vector')
plt.text(transformed_vector[0], transformed_vector[1], f'({transformed_vector[0]}, {transformed_vector[1]})', verticalalignment='bottom',
         horizontalalignment='right', fontsize=12, color='r')

plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Original Vector')
plt.text(vector[0], vector[1], f'({vector[0]}, {vector[1]})', verticalalignment='bottom', horizontalalignment='right',
         fontsize=12, color='b')

# Plot the angle between the vectors
plt.annotate('', xy=(transformed_vector[0], transformed_vector[1]), xytext=(vector[0], vector[1]))

plt.text(1, 2, f'$\\theta={theta_degrees:.2f}^{{\\circ}}$', fontsize=12)  # Display angle with theta symbol

plt.xlim(0, 20)
plt.ylim(0, 18)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Rotated and Scaled')

plt.grid()

# Customize the x-axis ticks
plt.xticks(np.arange(0, 21, 3))
plt.yticks(np.arange(0, 18, 4))

plt.legend(loc='upper left')

plt.show()
```

## 5. Eigen values and Eigen vectors determined using numpy functions

```python
import numpy as np

# Define the matrix A
A = np.array([[2, 3],
              [4, 1]])

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:")
print(eigenvectors)
```

## 7. Projection of matrix onto eigen vectors Calculation

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[2, 3],
              [4, 1]])

# Define the eigenvectors 
#v1 = np.array([1, 1])
#v2 = np.array([-3, 4])

# Define the eigenvectors 
v1 = np.array([5, 5])
v2 = np.array([6, -8])


# Scalar projections of the vectors in A onto v1 and v2
scalar_proj_Av1 = np.dot(A, v1) / np.linalg.norm(v1)
scalar_proj_Av2 = np.dot(A, v2) / np.linalg.norm(v2)

# Compute the projections of the vectors in A onto Av1 and Av2
projection_Av1 = scalar_proj_Av1 * (v1 / np.linalg.norm(v1))
projection_Av2 = scalar_proj_Av2 * (v2 / np.linalg.norm(v2))

print("projection_Av1:", projection_Av1)
print("projection_Av2:", projection_Av2)

```

<h5>The projection_Av1 has the highest value implying that eigen vector 1 has captured most important features of the matrix </h5>

#### Magnitude and Angle between vectors

```python
import numpy as np

# Define the vectors
a = np.array([1, 2])
b = np.array([5, 1])

# Compute the dot product
dot_product = np.dot(a, b)

# Compute the magnitudes of the vectors
magnitude_a = np.linalg.norm(a)
magnitude_b = np.linalg.norm(b)

# Compute the cosine of the angle between the vectors
cos_theta = dot_product / (magnitude_a * magnitude_b)

# Compute the angle in radians
theta_radians = np.arccos(cos_theta)

# Convert radians to degrees
theta_degrees = np.degrees(theta_radians)

print("Vector dot product (a,b):", dot_product)

print("Angle between vectors (in radians):", theta_radians)
print("Angle between vectors (in degrees):", theta_degrees)

print("magnitude_a:", magnitude_a)
print("magnitude_b:", magnitude_b)
```
